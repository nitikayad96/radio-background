# Statistical Tests Trials#

This notebook will be used for testing of different statistical tests to be used for evaluating the Gaussianity of the noise in residuals for each model

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from matplotlib.ticker import FormatStrFormatter
```

```python
ks100 = []
ks1000 = []
ks10000 = []

D100 = []
D1000 = []
D10000 = []

shapiro100 = []
shapiro1000 = []
shapiro10000 = []

W100 = []
W1000 = []
W10000 = []


for i in range(10000):

    # make a set of random numbers
    norm10000 = np.random.randn(10000)
    norm1000 = norm10000[0::10]
    norm100 = norm1000[0::10]

    # find KS test p-values
    ks100.append(stats.kstest(norm100, 'norm')[1])
    ks1000.append(stats.kstest(norm1000, 'norm')[1])
    ks10000.append(stats.kstest(norm10000, 'norm')[1])
    
    # find KS test statistic-values
    D100.append(stats.kstest(norm100, 'norm')[0])
    D1000.append(stats.kstest(norm1000, 'norm')[0])
    D10000.append(stats.kstest(norm10000, 'norm')[0])
    
    # find Shapiro Wilk p-values
    shapiro100.append(stats.shapiro(norm100)[1])
    shapiro1000.append(stats.shapiro(norm1000)[1])
    shapiro10000.append(stats.shapiro(norm10000)[1])
    
    # find Shapiro Wilk statistic values
    W100.append(stats.shapiro(norm100)[0])
    W1000.append(stats.shapiro(norm1000)[0])
    W10000.append(stats.shapiro(norm10000)[0])
```

```python
# plot p-value distributions

fig, axes = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(15,4))

ax1 = plt.subplot(1,3,1)
ax1.hist(ks100, bins=50, color='r')
ax1.set_title('100 samples')
plt.ylabel('Number of occurrences')

ax2 = plt.subplot(1,3,2)
ax2.hist(ks1000, bins=50, color='b')
ax2.set_title('1000 samples')
plt.xlabel('KS test p-value')

ax3 = plt.subplot(1,3,3)
ax3.hist(ks10000, bins=50, color='g')
ax3.set_title('10000 samples')

#plt.savefig('notes/kstest_pvalues.jpg')
```

```python
# plot p-value distributions
fig, axes = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(15,4))

ax1 = plt.subplot(1,3,1)
ax1.hist(np.array(D100)*np.sqrt(100), bins=50, color='r')
ax1.set_title('100 samples')
plt.ylabel('Number of occurrences')

ax2 = plt.subplot(1,3,2)
ax2.hist(np.array(D1000)*np.sqrt(1000), bins=50, color='b')
ax2.set_title('1000 samples')
plt.xlabel('KS test D-value')

ax3 = plt.subplot(1,3,3)
ax3.hist(np.array(D10000)*np.sqrt(10000), bins=50, color='g')
ax3.set_title('10000 samples')
#ax3.xaxis.set_major_formatter(FormatStrFormatter('%0.4f'))
#ax3.xaxis.set_major_locator(plt.MaxNLocator(5))

#plt.savefig('notes/kstest_Dvalues.jpg')
```

```python
plt.figure()
plt.plot(np.array(D100)*np.sqrt(100),ks100, 'r.')
plt.plot(np.array(D1000)*np.sqrt(1000),ks1000,'b.')
plt.plot(np.array(D10000)*np.sqrt(10000),ks10000, 'g.')
```

```python
# plot p-value distributions

fig, axes = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(15,4))

ax1 = plt.subplot(1,3,1)
ax1.hist(shapiro100, bins=50, color='r')
ax1.set_title('100 samples')
plt.ylabel('Number of occurrences')

ax2 = plt.subplot(1,3,2)
ax2.hist(shapiro1000, bins=50, color='b')
ax2.set_title('1000 samples')
plt.xlabel('Shapiro Wilk test p-value')

ax3 = plt.subplot(1,3,3)
ax3.hist(shapiro10000, bins=50, color='g')
ax3.set_title('10000 samples')

#plt.savefig('notes/shapirotest_pvalues.jpg')
```
