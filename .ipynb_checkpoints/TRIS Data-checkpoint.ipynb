```python
# Load two lowest bands from TRIS data, 0.6 GHz and 0.8 GHz

TRIS600txt = np.loadtxt('TRIS_absolute_600.txt', dtype=str)
TRIS600_ra = TRIS600txt[:,0]
TRIS600_Tb = np.float_(TRIS600txt[:,1])
TRIS600_Tbsig = np.float_(TRIS600txt[:,2])

TRIS820txt = np.loadtxt('TRIS_absolute_820.txt', dtype=str)
TRIS820_ra = TRIS820txt[:,0]
TRIS820_Tb = np.float_(TRIS820txt[:,1])
TRIS820_Tbsig = np.float_(TRIS820txt[:,2])
```

```python
# Create a SkyCoord object to turn RA and Dec into galactic coords

sc600 = SkyCoord(ra=TRIS600_ra, dec=42.0, unit=(u.hourangle,u.deg), frame='icrs')
sc820 = SkyCoord(ra=TRIS820_ra, dec=42.0, unit=(u.hourangle,u.deg))

sc600_gal = sc600.galactic
sc600_idx = hp.ang2pix(NSIDE_dg, sc600_gal.l.value, sc600_gal.b.value, lonlat=True)
```

```python
#### Plot brightness temperature as function of RA for both bands

plt.plot(sc600.ra, TRIS600_Tb, label='0.6 GHz')
#plt.plot(sc820.ra, TRIS820_Tb, label='0.8 GHz')
plt.title('TRIS Measurements', fontsize=12)
plt.xlabel('RA', fontsize=12)
plt.ylabel('Brightness Temperature (K)', fontsize=12)
plt.legend()
#plt.savefig('notes/TRIS_Tb.jpg')

idxmax = np.argmax(TRIS600_Tb)
idxrange = np.arange(idxmax-3, idxmax+5)
TRIS600mask = np.array(TRIS600_Tb)
TRIS600mask[idxrange]=0

idxmask = np.where(TRIS600mask != 0)

plt.plot(sc600.ra[idxmask], TRIS600_Tb[idxmask], 'g.')
plt.plot()

```

```python
# Create a healpy map showing TRIS measurement locations

map_1420 = hp.read_map('STOCKERT+VILLA-ELISA_1420MHz_1_256.fits')/1000
map_1420_dg = hp.pixelfunc.ud_grade(map_1420, 32)
map_1420_dg[np.where(map_1420_dg<0)] = None
map_1420_dg[sc600_idx[idxmask]] = None

hp.mollview(map_1420_dg, max=5)
hp.graticule()
```

```python
# Calcualte galactic temperature for 0.6 GHz given two reference positions and plot
aref1 = np.ravel(np.where(TRIS820_ra=='10h00m'))
Tgal_ref1 = 2.21
Tsky_ref1 = TRIS820_Tb[aref1]
Tgal1 = TRIS820_Tb - Tsky_ref1 + Tgal_ref1

aref2 = np.ravel(np.where(TRIS820_ra=='20h23m'))
Tgal_ref2 = 10.38
Tsky_ref2 = TRIS820_Tb[aref2]
Tgal2 = TRIS820_Tb - Tsky_ref2 + Tgal_ref2

plt.plot(sc820.ra, TRIS820_Tb, label='$T_{sky}$')
plt.plot(sc820.ra, Tgal1,'-', label=('$T_{gal}$' + ' from RA=10h00m'))
plt.plot(sc820.ra, Tgal2,'--', label=('$T_{gal}$' + ' from RA=20h24m'))
plt.title('TRIS Galactic Temperature at 0.82 GHz', fontsize=12)
plt.xlabel('RA', fontsize=12)
plt.ylabel('Brightness Temperature (K)', fontsize=12)
plt.legend();
#plt.savefig('notes/TRIS_Tgal.jpg')
```

## Visualize MCMC results ##
These MCMC results seek to find the largest halo that would be consistent with the data. This can later be used to compare the result with the EDGES survey results.

```python
samples = np.load("TRIS_sph+halo.npz")['samples']
samples = samples[:,500:,:]
print(samples.shape)

ndim=samples.shape[2]
samples_reshape = samples.reshape((-1, ndim))
print(samples_reshape.shape)

samples_reshape[:,0] = np.log10(samples_reshape[:,0])
samples_reshape[:,1] = np.log10(samples_reshape[:,1])
samples_reshape[:,2] = np.log10(samples_reshape[:,2])
samples_reshape[:,3] = np.log10(samples_reshape[:,3])
samples_reshape[:,4] = np.log10(samples_reshape[:,4])

# if np.any(np.isnan(samples_reshape)==True):
#     print("there are nans")
```

```python
R_disk = 10**np.percentile(samples_reshape[:,0], 50)
h_disk = 10**np.percentile(samples_reshape[:,1], 50)
j_disk = 10**np.percentile(samples_reshape[:,2], 50)
R_halo = 10**np.percentile(samples_reshape[:,3], 50)
# h_halo = 10**np.percentile(samples_reshape[:,4], 50)
j_halo = 10**np.percentile(samples_reshape[:,4], 50)

print(R_disk/d)
print(h_disk/(1000*pc))
print(R_halo/d)


nu1 = 0.6e9
#nu2 = 0.82e9

disk = MD.Spheroid(TRIS_l[nu1], TRIS_b[nu1], R_disk, h_disk)*j_disk
halo = MD.LineOfSightHalo(TRIS_l[nu1], TRIS_b[nu1], d, R_halo)*j_halo
model = (disk+halo)*(c**2)/(2*k*(nu1**2)) 

residuals = TRIS_Tgal[nu1] - model

plt.figure()
plt.plot(sc600.ra, TRIS_Tgal[nu1], '-', label='0.6 GHz')
plt.plot(sc600.ra, model, '-',label='Model')
plt.plot(sc600.ra, residuals, '-', label='Residuals')
plt.title('Fitting TRIS Measurements with Spheroidal Disk Model', fontsize=12)
plt.xlabel('RA', fontsize=12)
plt.ylabel('Brightness Temperature (K)', fontsize=12)
plt.legend(loc='upper left');
plt.ylim([-7,27]);
# plt.savefig('notes/residuals_TRIS_sph.jpg')

# plt.figure()
# plt.plot(sc820.ra, TRIS_Tgal[nu2], label='0.82 GHz')
# plt.plot(sc820.ra, model2, label='Model')
# plt.plot(sc820.ra, res2, label='Residuals')
# plt.title('Fitting Multi-Frequency TRIS Data with Sph+Halo Model', fontsize=12)
# plt.xlabel('RA', fontsize=12)
# plt.ylabel('Brightness Temperature (K)', fontsize=12)
# plt.legend(loc='upper left');
# plt.ylim([-7,27]);
# plt.savefig('notes/residuals_TRIS_multi820.jpg')


# print(R_disk/d)
# print(h_disk/d)
# print(R_halo/d)
# print(h_disk/d)
```

## Make images to compare to CHANG-ES images ##

```python
r_halo = R_halo/(10*1e6*pc)*u.rad
r_halo = r_halo.to(u.arcmin)
print(r_halo)

n = 1000
npix=n**2
arcmin2 = 12**2
npix_arcmin2 = npix/arcmin2
arcmin2_beam = 0.6**2
perbeam = arcmin2_beam/arcmin2
x = np.linspace(-6,6,n)
X = np.tile(x, (n,1))
Y = X.T

r = np.sqrt(X**2 + Y**2)

sigma, mu = 0.36/2, 0.0 #beam of 0.36 arcmin FWHM
beam = np.exp(-( (r-mu)**2 / ( 2.0 * sigma**2 ) ) )
fig, ax = plt.subplots(1,1)
img = ax.imshow(beam, cmap='binary', extent = [-6 , 6, -6 , 6])

print(np.max(r))
check = r <= r_halo.value
D_halo = check*2*R_halo*np.cos(0.5*np.pi*r/r_halo.value)*j_halo*1000/(npix*Jy)    #*(c**2)/(2*k*(nu1**2))
D_halo[np.where(np.isnan(D_halo))] = 0


fig, ax = plt.subplots(1,1)
img = ax.imshow(D_halo, cmap='binary', extent = [-6 , 6, -6 , 6])
ax.set_xticks([])
ax.set_yticks([-5,0,5])
y_label_list = ['-05\'', '+00\'', '+05\'']
ax.set_yticklabels(y_label_list)
cbar = fig.colorbar(img)
cbar.ax.get_xaxis().labelpad = 10
cbar.ax.set_xlabel('mJy/pixel', fontsize=12)
plt.title('Simulated Galaxy Halo at 10 Mpc');
# plt.savefig('notes/simgal.jpg')
```

```python
from scipy import ndimage, signal

r_disk = R_disk/(10*1e6*pc)*u.rad
r_disk = r_disk.to(u.arcmin).value

h_disk_ = h_disk/(10*1e6*pc)*u.rad
h_disk_ = h_disk_.to(u.arcmin).value

check2 = ((X/r_disk)**2 + (Y/h_disk_)**2) <= 1
Y_ = h_disk_*np.sqrt(1 - (X/r_disk)**2)
D_sph = check2*2*R_disk*np.cos(0.5*np.pi*X/r_disk)*np.sqrt(1 - (Y/Y_)**2)*j_disk*1000/(npix*Jy)
D_sph[np.where(np.isnan(D_sph))] = 0

D_tot = D_halo+D_sph
np.savez("simgal_disk+halo.npz", data=D_tot)
np.savez("simgal_disk.npz", data=D_sph)
np.savez("simgal_halo.npz", data=D_halo)

fig, ax = plt.subplots(1,1)
img = ax.imshow(D_tot, cmap='binary', extent = [-6 , 6, -6 , 6], vmax=100)
ax.set_xticks([])
ax.set_yticks([-5,0,5])
y_label_list = ['-05\'', '+00\'', '+05\'']
ax.set_yticklabels(y_label_list)
cbar = fig.colorbar(img)
cbar.ax.get_xaxis().labelpad = 10
cbar.ax.set_xlabel('mJy/pixel', fontsize=12);
plt.title('Simulated Galaxy Disk+Halo at 10 Mpc');
#plt.savefig('notes/simgal2.jpg')

final_img = signal.convolve2d(D_tot, beam, 'same')
#final_img = ndimage.convolve(D_tot, beam, mode='nearest', cval=0.0)
fig, ax = plt.subplots(1,1)
img = ax.imshow(final_img, cmap='binary', extent = [-6 , 6, -6 , 6])
```
